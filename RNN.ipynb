```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import mean_squared_error, mean_absolute_error

from statsmodels.tsa.seasonal import seasonal_decompose

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import SimpleRNN, LSTM, GRU, Dense

# 2. LOAD DATASET
df = pd.read_csv("monthly_milk_production.csv")

print(df.head())

# 3. DATA PREPROCESSING

# Rename columns if needed
df.columns = ['Date', 'Production']

# Convert Date column
df['Date'] = pd.to_datetime(df['Date'])

# Set Date as index
df.set_index('Date', inplace=True)

print(df.info())

# Check missing values
print("\nMissing Values:")
print(df.isnull().sum())

# Summary statistics
print("\nSummary Statistics:")
print(df.describe())

# 4. EXPLORATORY DATA ANALYSIS

# Time Series Plot
plt.figure(figsize=(12,5))
plt.plot(df.index, df['Production'])
plt.title('Monthly Milk Production')
plt.xlabel('Date')
plt.ylabel('Production')
plt.grid(True)
plt.show()

# Boxplot for outliers
plt.figure(figsize=(6,4))
sns.boxplot(x=df['Production'])
plt.title('Boxplot of Production')
plt.show()

# Seasonal Decomposition
result = seasonal_decompose(df['Production'],
                            model='additive',
                            period=12)

result.plot()
plt.show()

# 5. NORMALIZATION

scaler = MinMaxScaler()

scaled_data = scaler.fit_transform(df[['Production']])

# 6. CREATE SEQUENCES

def create_sequences(data, window_size):

    X = []
    y = []

    for i in range(len(data)-window_size):

        X.append(data[i:i+window_size])
        y.append(data[i+window_size])

    return np.array(X), np.array(y)

window_size = 12

X, y = create_sequences(scaled_data, window_size)

print("X Shape:", X.shape)
print("y Shape:", y.shape)

# 7. TRAIN TEST SPLIT

train_size = int(len(X) * 0.8)

X_train = X[:train_size]
X_test = X[train_size:]

y_train = y[:train_size]
y_test = y[train_size:]

# 8. BUILD RNN MODEL

rnn_model = Sequential([
    SimpleRNN(50,
              activation='tanh',
              input_shape=(window_size,1)),
    Dense(1)
])

rnn_model.compile(
    optimizer='adam',
    loss='mse'
)

print("\nRNN MODEL SUMMARY")
rnn_model.summary()

# Train RNN
history_rnn = rnn_model.fit(
    X_train,
    y_train,
    epochs=50,
    batch_size=16,
    validation_split=0.1,
    verbose=1
)

# 9. BUILD LSTM MODEL

lstm_model = Sequential([
    LSTM(50,
         input_shape=(window_size,1)),
    Dense(1)
])

lstm_model.compile(
    optimizer='adam',
    loss='mse'
)

print("\nLSTM MODEL SUMMARY")
lstm_model.summary()

# Train LSTM
history_lstm = lstm_model.fit(
    X_train,
    y_train,
    epochs=50,
    batch_size=16,
    validation_split=0.1,
    verbose=1
)

# 10. BUILD GRU MODEL
gru_model = Sequential([
    GRU(50,
        input_shape=(window_size,1)),
    Dense(1)
])

gru_model.compile(
    optimizer='adam',
    loss='mse'
)

print("\nGRU MODEL SUMMARY")
gru_model.summary()

# Train GRU
history_gru = gru_model.fit(
    X_train,
    y_train,
    epochs=50,
    batch_size=16,
    validation_split=0.1,
    verbose=1
)

# 11. PREDICTIONS

rnn_pred = rnn_model.predict(X_test)
lstm_pred = lstm_model.predict(X_test)
gru_pred = gru_model.predict(X_test)

# Inverse transform
rnn_pred = scaler.inverse_transform(rnn_pred)
lstm_pred = scaler.inverse_transform(lstm_pred)
gru_pred = scaler.inverse_transform(gru_pred)

y_actual = scaler.inverse_transform(y_test)

# 12. EVALUATION METRICS

def evaluate(actual, pred):

    rmse = np.sqrt(mean_squared_error(actual, pred))

    mae = mean_absolute_error(actual, pred)

    mape = np.mean(
        np.abs((actual - pred) / actual)
    ) * 100

    return rmse, mae, mape

rnn_scores = evaluate(y_actual, rnn_pred)
lstm_scores = evaluate(y_actual, lstm_pred)
gru_scores = evaluate(y_actual, gru_pred)

print("\n==============================")
print("MODEL PERFORMANCE")
print("==============================")

print(f"RNN  -> RMSE: {rnn_scores[0]:.2f}, MAE: {rnn_scores[1]:.2f}, MAPE: {rnn_scores[2]:.2f}%")

print(f"LSTM -> RMSE: {lstm_scores[0]:.2f}, MAE: {lstm_scores[1]:.2f}, MAPE: {lstm_scores[2]:.2f}%")

print(f"GRU  -> RMSE: {gru_scores[0]:.2f}, MAE: {gru_scores[1]:.2f}, MAPE: {gru_scores[2]:.2f}%")

# 13. PLOT ACTUAL VS PREDICTED

# RNN Plot
plt.figure(figsize=(12,5))
plt.plot(y_actual, label='Actual')
plt.plot(rnn_pred, label='RNN Prediction')
plt.title('RNN Predictions')
plt.legend()
plt.show()

# LSTM Plot
plt.figure(figsize=(12,5))
plt.plot(y_actual, label='Actual')
plt.plot(lstm_pred, label='LSTM Prediction')
plt.title('LSTM Predictions')
plt.legend()
plt.show()

# GRU Plot
plt.figure(figsize=(12,5))
plt.plot(y_actual, label='Actual')
plt.plot(gru_pred, label='GRU Prediction')
plt.title('GRU Predictions')
plt.legend()
plt.show()

# 14. FORECAST NEXT 12 MONTHS

future_input = scaled_data[-window_size:]
future_input = future_input.reshape(1, window_size, 1)

future_predictions = []

# Using LSTM for forecasting
for i in range(12):

    pred = lstm_model.predict(
        future_input,
        verbose=0
    )

    future_predictions.append(pred[0,0])

    future_input = np.append(
        future_input[:,1:,:],
        [[[pred[0,0]]]],
        axis=1
    )

# Convert back to original scale
future_predictions = np.array(
    future_predictions
).reshape(-1,1)

future_predictions = scaler.inverse_transform(
    future_predictions
)

print("\nNEXT 12 MONTH FORECAST")
print(future_predictions)

# 15. PLOT FUTURE FORECAST

future_dates = pd.date_range(
    start=df.index[-1] + pd.DateOffset(months=1),
    periods=12,
    freq='MS'
)

plt.figure(figsize=(12,5))

plt.plot(
    df.index,
    df['Production'],
    label='Historical Data'
)

plt.plot(
    future_dates,
    future_predictions,
    label='Future Forecast',
    color='red'
)

plt.title('12-Month Milk Production Forecast')
plt.xlabel('Date')
plt.ylabel('Production')

plt.legend()
plt.grid(True)

plt.show()

# 16. FINAL COMPARISON TABLE

results = pd.DataFrame({
    'Model': ['RNN', 'LSTM', 'GRU'],
    'RMSE': [rnn_scores[0], lstm_scores[0], gru_scores[0]],
    'MAE': [rnn_scores[1], lstm_scores[1], gru_scores[1]],
    'MAPE': [rnn_scores[2], lstm_scores[2], gru_scores[2]]
})

print("\nFINAL MODEL COMPARISON")
print(results)
```
